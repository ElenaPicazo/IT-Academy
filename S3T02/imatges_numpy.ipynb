### Exercici 3
Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
def llista_aleatoria(nums, num1,num2):
    randomlist = []
    for i in range(nums):
        n = random.randint(num1,num2)
        randomlist.append(n)
    return randomlist

# crear 5 llistes perquè volem una matriu amb 5 files i cada fila amb 5 columnes
final_lst=[]
for i in range(5):
    row=llista_aleatoria(5,1,9)
    final_lst.append(row)
```

```python
array_2D=np.array(final_lst, dtype=np.int64)

# Valor màxim
val_max=np.max(array_2D)

# valors màxims de cadascun dels seus eixos.
## valor màxim de les files: 
max_row=0
for row in array_2D:
    val_max=np.max(row)
    if val_max>max_row:
        max_row=val_max
    
print(max_row)
```

```python
## valor màxim de les columnes: 
lst_maxs_cols=[]
for col in array_2D.T:
    maxi=np.max(col)
    lst_maxs_cols.append(maxi)
    
max_cols=max(lst_maxs_cols)
max_cols
```

Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.
